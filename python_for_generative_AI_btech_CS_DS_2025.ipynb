```python
#while loop
i = 0 #inilization
while i<5: #condition
  print(i)
  i+=1 #increment
```

```python
#loops and control statement
#break -- this statement break the loop
i = 0
while i<5:
  i+=1 #increment
  if i==3:
    break
  else:
    print(i)
```

```python
#continue statement-- this statement skip the current iteration
i = 0
while i<5:
  i+=1
  if i==3:
    continue
    print(i)
  else:
    print(i)

```

```python
#pass statement -- this statement pass the code
i = 0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```

```python
#default argument
def area_perimeter(width=4, height=8):
  area = width*height
  perimeter = 2*(width + height)
  return "Area is", area, "and perimeter is", perimeter
```

```python
print(area_perimeter(10, 20))
```

```python
#keyword argument
def interest(p, r, t):
  i = (p*r*t)/100
  return i
```

```python
print(interest(r = 10, t = 2, p = 1000))
```

```python
#mixed argument
#positional argument follows keyword argument
print(interest(1000, t = 2, r = 10))
```

```python
print(interest(r = 10,p=1000,  2))
```

```python
#variable length argument
#non keyword positional variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2,3, 4, 8, 9, 7)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(3, 5, 6, 7, 8, 9)
```

```python
#keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(type(kwargs))
  print(len(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
def pass_students(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_students(A = 30, B = 40, C = 50, D= 80, E = 20)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 80, 50])
color = np.array(["red", "yellow", "orange", "green"])
plt.bar(fruit, quantity, color = color)
plt.xlabel("Fruit")
plt.ylabel("Quantity")
plt.title("Fruit Quantity")
plt.show()
```

```python
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 80, 50])
color = np.array(["red", "yellow", "orange", "green"])
plt.barh(fruit, quantity, color = color)
plt.xlabel("Quantity")
plt.ylabel("Fruit")
plt.title("Fruit Quantity")
plt.show()
```

```python
#pie plot
quantity = np.array([20, 35, 15, 30])
fruit = np.array(["Apple", "Banana", "Orange", "Mango"])
myexplode = [0.2, 0, 0, 0]
plt.pie(quantity, labels = fruit, explode = myexplode)
plt.legend(title = "Fruit",loc = "best")
plt.show()
```
