```python
#while
i = 0 #initialization
while i<5:
  print(i)
  i+=1
```

```python
#break--this statement break the loop
i = 0
while i<5:
  i+=1
  if i == 3:
    break
  else:
    print(i)
```

```python
#continue statement--skip the current iteration
i = 0
while i<5:
  i+=1
  if i == 3:
    continue
    print(i)
  else:
    print(i)
```

```python
#pass statement---pass the code
i = 0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```

```python
#default argument
def area_perimeter(width = 4, height = 8):
  area = width*height
  perimeter = 2*(width+height)
  return "Area is", area, "and perimeter is ", perimeter
```

```python
area_perimeter(10, 20)
```

```python
#keyword argument
def interest(p, r, t):
  i = (p*r*t)/100
  return i
```

```python
#bar plot--it is work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 90, 80])
color = np.array(["red", "yellow", "orange", "green"])
plt.bar(fruit, quantity, color = color)
plt.show()
```

```python
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 90, 80])
color = np.array(["red", "yellow", "orange", "green"])
plt.barh(fruit, quantity, color = color)
plt.show()
```

```python
#pie plot
fruit_percentage = np.array([20, 25, 35, 20])
fruit_name = np.array(["Apple", "Banana", "Orange", "Watermelon"])
colors = np.array(["red", "yellow", "orange", "green"])
myexplode = np.array([0.2, 0, 0, 0])
plt.pie(fruit_percentage, labels = fruit_name, autopct="%1.1f%%",
        colors =colors, explode = myexplode  )
plt.legend(title = "Fruit", loc = "best")
plt.show()
```

```python
age_of_car = np.array([4,5, 6, 7, 8, 9, 10, 11, 12])
speed_of_car = np.array([150, 140, 130, 120, 110, 100, 90, 80, 70])
plt.scatter(age_of_car, speed_of_car, color = "red")
plt.xlabel("Age of car")
plt.ylabel("Speed of car")
plt.title("Age vs speed of car")
plt.show()
```
