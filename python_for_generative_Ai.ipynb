```python
# Function to multiply numbers using non-keyword positional variable length argument
def multiply(*numbers):
  """Multiplies all the numbers passed as arguments."""
  result = 1  # Start with 1 so the product isn't zero
  for number in numbers:
    result *= number
  return result

# Example usage
print(multiply(2, 3, 4))
print(multiply(5, 10, 2, 0.5))
print(multiply(7))
```

```python
vimport pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Create Sample Data
# We'll create a DataFrame, which is like a spreadsheet in Python.
data = {
    'Size_sqft': [1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700],
    'Price_usd': [245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000, 282000, 299000]
}
df = pd.DataFrame(data)

# ---

# 2. Prepare Data and Train the Model
# X is the independent variable (input), y is the dependent variable (output).
X = df[['Size_sqft']]  # Needs to be a 2D array, so we use double brackets
y = df['Price_usd']

# Create an instance of the linear regression model
model = LinearRegression()

# Train the model using our data (this is where it finds the 'line of best fit')
model.fit(X, y)

# ---

# 3. Make a Prediction
# Let's predict the price for a house that is 2000 sq ft.
predicted_price = model.predict([[2000]])

print(f"Model Intercept (): {model.intercept_:.2f}")
print(f"Model Slope/Coefficient (): {model.coef_[0]:.2f}")
print("-" * 30)
print(f"Predicted price for a 2000 sq ft house: ${predicted_price[0]:.2f}")

# ---

# 4. Visualize the Results

# Plot the actual data points
plt.scatter(X, y, color='blue', label='Actual Prices')

# Plot the regression line (the model's predictions)
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Line of Best Fit')

# Add labels and a title
plt.title('House Price vs. Size')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
